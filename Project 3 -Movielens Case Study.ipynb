```python
# PERFORMING LOGISTIC REGRRESSION ON DATA - Master_Data
#  Target Column = Rating
#  Feature Columns = Genere,Age,Occupation
feature_columns = ['Age','Occupation']
x = Master_Data[feature_columns]
y = Master_Data.Rating

```

```python
# Splitting the Data as 75:25, into training and testing sets
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.40,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression(solver='lbfgs',multi_class='multinomial')
```

```python
# fit the model with data
logreg.fit(X_train,y_train)
```

```python
y_pred=logreg.predict(X_test)
```

```python

#Evaluating the Model
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
