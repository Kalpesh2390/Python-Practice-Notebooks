```python
# 3. Define a function to eliminate stopwords

def eliminate_stopwords(message):
    # Eliminate the Punctuations
    char_Wo_punctuations = [char for char in message if char not in string.punctuation]
    # join 
    Sentence_wo_Punctuation = ''.join(char_Wo_punctuations)
    # Eliminate the stopword and return the clean sentense
    Clean_setense = [word.lower() for word in Sentence_wo_Punctuation.split() 
                     if word.lower() not in stopwords.words('english')]
    return Clean_setense
    
```
